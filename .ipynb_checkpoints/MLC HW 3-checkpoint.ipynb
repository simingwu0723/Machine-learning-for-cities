### Q2. Bayesian Network Learning (35 points)

In this question, we use dataset: "HW3Q2.csv" for Bayesian Network Learning.

```python
from sklearn.model_selection import train_test_split
data2=pd.read_csv("HW3Q2.csv")
train,test=train_test_split(data2,random_state=9,test_size=0.4)
data2.head()
```

a) Use the training data to select the best structure you want to use for Bayesian Network Learning. Please use Hill Climbing with BIC score metric. (10 points)

b) Use the Bayesian Estimator to estimate the CPDs for your model and visualize the network with CPDs. (15 points)

c) Use the model to predict "A" for the testing dataset. Report the out-of-sample prediction accuracy. (10 points)   

```python
from pgmpy.estimators import BicScore
from pgmpy.estimators import HillClimbSearch
from pgmpy.models import BayesianModel
from pgmpy.estimators import BayesianEstimator
```

```python
# a)
hc = HillClimbSearch(train)
best_model = hc.estimate(scoring_method=BicScore(train))
print(best_model.edges())
```

```python
# b)
best_model = BayesianModel(best_model.edges())
best_model.fit(train, estimator=BayesianEstimator, prior_type='K2')
for cpd in best_model.get_cpds():
    print("CPD of {variable}:".format(variable=cpd.variable))
    print(cpd)
```

```python
# c)
predicted_test = best_model.predict(test.loc[:, ["B","D","E"]])
print("Out of sample:",(test.loc[:,"A"].reset_index(drop=True)==predicted_test["A"]).mean())
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import linkage, fcluster
from sklearn.mixture import GaussianMixture
```

```python
# g)

X3_day = data3.loc[:,"day_Sun":"day_Sat"]
range_n_clusters = range(2,10)
```

```python
# Kmeans

print("Running k-means:")
for n_clusters in range_n_clusters:
    silhouette = []
    for i in range(100):
        km = KMeans(random_state=i, n_clusters=n_clusters)
        cluster_labels = km.fit(X3_day).labels_
        silhouette.append(silhouette_score(X3_day, cluster_labels))
    silhouette_avg = np.mean(silhouette)
    print("For n_clusters = {},".format(n_clusters)+" the average silhouette_score is :{}".format(silhouette_avg))
```

```python
# For K-means, choose n_clusters = 5.

print("For 5 clusters:")
km_best = KMeans(random_state=1, n_clusters=5)
km_best_labels = km_best.fit(X3_day).labels_
for i in range(5):
    print('Number of elements in cluster {} : {}'.format(i,np.sum(km_best_labels==i)))
```

```python
# Gaussian mixture models

print("Running Gaussian mixture models:")
for n_clusters in range_n_clusters:
    silhouette = []
    for i in range(100):
        gm = GaussianMixture(random_state=i, n_components=n_clusters)
        cluster_labels = gm.fit(X3_day).predict(X3_day)
        silhouette.append(silhouette_score(X3_day, cluster_labels))
    silhouette_avg = np.mean(silhouette)
    print("For n_clusters = {},".format(n_clusters)+" the average silhouette_score is :{}".format(silhouette_avg))
```

```python
# For Gaussian mixture model, choose n_clusters = 2

print("For 2 clusters:")
gm_best = GaussianMixture(random_state=1, n_components=2)
gm_best_labels = gm_best.fit(X3_day).predict(X3_day)
for i in range(2):
    print('Number of elements in cluster {} : {}'.format(i,np.sum(gm_best_labels==i)))
```

```python
#  Bottom-up hierarchical clustering with "single link" distance metric

print("Running hierarchical single linkage:")
hc_single = linkage(X3_day, 'single')
for n_clusters in range_n_clusters:
    cluster_labels = fcluster(hc_single, n_clusters, criterion='maxclust')  
    silhouette_avg = silhouette_score(X3_day, cluster_labels)
    print("For n_clusters = {},".format(n_clusters)+" the average silhouette_score is :{}".format(silhouette_avg))
```

```python
# For bottom-up hierarchical clustering with "single link" distance metric, choose n_clusters = 2

print("For 2 clusters:")
hc_best_labels = fcluster(hc_single, 2, criterion='maxclust')
for i in range(1, 3):
    print('Number of elements in cluster {} : {}'.format(i,np.sum(hc_best_labels==i)))
```

```python
# Bottom-up hierarchical clustering with "complete link" distance metric

print("Running hierarchical complete linkage:")
hc_complex = linkage(X3_day, 'complete')
for n_clusters in range_n_clusters:
    cluster_labels = fcluster(hc_complex, n_clusters, criterion='maxclust')  
    silhouette_avg = silhouette_score(X3_day, cluster_labels)
    print("For n_clusters = {},".format(n_clusters)+" the average silhouette_score is :{}".format(silhouette_avg))
```

```python
# For bottom-up hierarchical clustering with "complete link" distance metric, choose n_clusters = 2

print("For 2 clusters:")
hc_best_labels = fcluster(hc_complex, 2, criterion='maxclust')
for i in range(1,3):
    print('Number of elements in cluster {} : {}'.format(i,np.sum(hc_best_labels==i)))
```

The cluster distributions generated by K-means and Gaussian mixture modeling are less biased compared to both hierarchical clustering methods, with a more even distribution of samples among the clusters.
